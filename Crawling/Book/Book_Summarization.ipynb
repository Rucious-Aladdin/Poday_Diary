```python
import openai
import pandas as pd
import time

# OpenAI API 키 설정
openai.api_key = ''

# 책 데이터 불러오기
file_path = '/content/test.csv'
df = pd.read_csv(file_path, encoding='utf-8')

# 책의 줄거리 추출 함수
def extract_summary(introduction):
    try:
        response = openai.ChatCompletion.create(
            model="gpt-4",
            messages=[
                {"role": "system", "content": "You are a helpful assistant."},
                {"role": "user", "content": f"다음은 책의 출판사 소개입니다. 이 소개에서 줄거리만 3줄로 요약해주세요:\n\n{introduction}\n\n줄거리:"}
            ],
            max_tokens=500,
            temperature=0.5,
            n=1,
            stop=None
        )
        summary = response['choices'][0]['message']['content'].strip()
        return summary
    except Exception as e:
        print(f"Error: {e}")
        return ""

# 책의 줄거리 추출 및 새로운 열에 저장
summaries = []
for intro in df['introduction']:
    summaries.append(extract_summary(intro))
    time.sleep(2)  # 각 요청 사이에 2초의 지연 시간을 추가합니다.

df['introduction2'] = summaries

# 결과를 CSV 파일로 저장
output_path = 'list_with_summary2.csv'
df.to_csv(output_path, index=False, encoding='utf-8')

print("요약 완료 및 저장 완료")
```

```python
def find_encoding_errors(file_path, encoding='utf-8'):
    with open(file_path, 'rb') as f:
        for i, line in enumerate(f):
            try:
                line.decode(encoding)
            except UnicodeDecodeError as e:
                print(f"Decoding error at line {i + 1}, position {e.start}: {e}")
                return e

file_path = '/content/test5.csv'
error = find_encoding_errors(file_path)

if error:
    print("문제를 일으키는 문자 위치와 원인:")
    print(error)
else:
    print("파일을 성공적으로 읽었습니다. 인코딩 오류가 없습니다.")
```
