```python
def select_columns(df, columns):
  return df[columns]
```

```python
year4_filename = "4차년도.csv"
year5_filename1 = "5차년도.csv"
year5_filename2 = "5차년도_2차.csv"
columns = ["발화문", "1번 감정", "2번 감정", "3번 감정", "4번 감정", "5번 감정" ]
year4_df = pd.read_csv(os.path.join(path, year4_filename), sep=",", encoding="cp949")
year5_df1 = pd.read_csv(os.path.join(path, year5_filename1), sep=",", encoding="cp949")
year5_df2 = pd.read_csv(os.path.join(path, year5_filename2), sep=",", encoding="cp949")
year4_df = select_columns(year4_df, columns)
year5_df1 = select_columns(year5_df1, columns)
year5_df2 = select_columns(year5_df2, columns)
```

```python
year4_df.head()
```

```python
df = pd.concat([year4_df, year5_df1, year5_df2], axis=0)
df.shape
```

```python
lower = lambda x: x.lower()
for col in ["1번 감정", "2번 감정", "3번 감정", "4번 감정", "5번 감정"]:
    df[col] = df[col].apply(lower)
```

```python
# Function to perform majority voting with tie resolution
def majority_vote(row):
    emotions = row[1:]
    emotion_counts = Counter(emotions)
    most_common_emotions = emotion_counts.most_common()
    max_count = most_common_emotions[0][1]
    candidates = [emotion for emotion, count in most_common_emotions if count == max_count]

    if len(candidates) > 1:
        return "equal"
    else:
        return candidates[0]

df['감정'] = df.apply(majority_vote, axis=1)
```

```python
df = df[df["감정"] != "equal"]
```

```python
df = select_columns(df, ["발화문", "감정"])
```

```python
df.groupby("감정").count().plot(kind="bar")
```

```python
from sklearn.model_selection import train_test_split

not_sad_angry_df = df[(df["감정"] != "sadness") & (df["감정"] != "angry")]
sad_df = df[df["감정"] == "sadness"]
angry_df = df[df["감정"] == "angry"]
san_undersampled, _ = train_test_split(sad_df, test_size=0.7, random_state=42)
angry_undersampled, _ = train_test_split(angry_df, test_size=0.5, random_state=42)
```

```python
df = pd.concat([san_undersampled, angry_undersampled, not_sad_angry_df], axis=0)
df.groupby("감정").count().plot(kind="bar")
```

```python
en_emo2int ={
    "joy": 0, "surprise": 1, "anger": 2,
    "fear": 3, "hurt": 4, "sadness": 5,
    "neutral": 6
}
```

```python
dis_to_angry = lambda  x: x if x != "disgust" else "angry"
hap_to_joy = lambda x: x if x != "happiness" else "joy"
add_usr_token = lambda x : "[USR] " + x
df["감정"] = df["감정"].apply(dis_to_angry)
df["감정"] = df["감정"].apply(hap_to_joy)
df["발화문"] = df["발화문"].apply(add_usr_token)
```

```python
df.columns = ["text", "label_name"]
```

```python
anger_to_angry = lambda  x: x if x != "angry" else "anger"
df["label_name"] = df["label_name"].apply(anger_to_angry)
label_name_to_int = lambda x : en_emo2int[x]
df["label"] = df["label_name"].apply(label_name_to_int)
```

```python
kai_train_df, kai_test_df = train_test_split(df, test_size=0.1, random_state=42)
kai_train_df, kai_valid_df = train_test_split(kai_train_df, test_size=0.2, random_state=42)
```

```python
kai_train_df.shape, kai_valid_df.shape, kai_test_df.shape
```

```python
kai_train_df = select_columns(kai_train_df, ["text", "label"])
kai_valid_df = select_columns(kai_valid_df, ["text", "label"])
kai_test_df = select_columns(kai_test_df, ["text", "label"])
```
